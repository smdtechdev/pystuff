```python
import requests
import os 
import pandas as pd
import numpy as np
import matplotlib 
import datetime
from fpl import FPL
from dotenv import load_dotenv
load_dotenv()

pd.options.display.max_columns = None
pd.options.mode.chained_assignment = None # default='warn'

user_name = os.environ.get('USER')
password = os.environ.get('PASS')
user_ID = os.environ.get('ID')
```

```python
stcurl = 'https://fantasy.premierleague.com/api/bootstrap-static/'
stcr = requests.get(stcurl)
stcjson = stcr.json()

player_info = pd.DataFrame(stcjson['elements'])
player_type = pd.DataFrame(stcjson['element_types'])
teams_info = pd.DataFrame(stcjson['teams'])
```

```python
session = requests.session()

url = 'https://users.premierleague.com/accounts/login/'

payload = {
 'password': password,
 'login': user_name,
 'redirect_uri': 'https://fantasy.premierleague.com/a/login',
 'app': 'plfpl-web'
}

session.post(url, data=payload)
```

```python
response = session.get('https://fantasy.premierleague.com/api/my-team/' + str(user_ID))
print(response)
json = response.json()
```
